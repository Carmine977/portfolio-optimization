## Optimize Portfolio

### Import libraries

```python
import pandas as pd
import numpy as np
import scipy.optimize as spo
from utils.util import get_data, plot_data, get_portfolio_value
from utils.util import compute_cumulative_returns, compute_daily_returns, compute_sharpe_ratio
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Initialization

```python
date_start = '2016-07-01'
date_end = '2017-07-08'

# Define portfolio composition
symbols = ['AAPL', 'ADBE', 'BA', 'BAC', 'GD', 'HRS', 'PCLN', 'PGR', 'STT', 'TWX']
symbols = ['PGR', 'BA', 'GD', 'STT', 'BAC']
#symbols = ['BA','BAC'] 
dates = pd.date_range(date_start, date_end)
```

### Load and plot stock data

```python
prices_all = get_data(symbols, dates, price='Close')
prices = prices_all[symbols]  
prices_SPY = prices_all['SPY']  # use SPY for comparison
plot_data(prices_all/prices_all.ix[0], title="Normalized Stock Prices")
```

### Optimize portfolio allocations

**Define some utility functions**

```python
def get_portfolio_stats(port_val, daily_rf=0, samples_per_year=252):
    """Calculates the cumulative return, average daily return (and standard deviation) and Sharpe ratio of a portfolio
    given the daily portfolio values, the daily risk-free rate of return, and the frequency of sampling 
    (or, how many trading days there are)"""
    # Compute cumulative ruturn 
    cum_ret = compute_cumulative_returns(port_val)
    # Compute daily returns, its mean and standard deviation
    daily_returns = compute_daily_returns(port_val)
    avg_daily_ret = daily_returns.mean()
    std_daily_ret = daily_returns.std()
    # calculate sharp ratio
    sharpe_ratio = compute_sharpe_ratio(port_val, daily_rf=0, samples_per_year=252)
    return cum_ret, avg_daily_ret, std_daily_ret, sharpe_ratio

def error_optimal_allocations(allocs, prices):
    """A helper function for the find_optimal_allocations function to minimize over"""
    port_val = get_portfolio_value(prices, allocs, 1)
    cum_ret, avg_daily_ret, std_daily_ret, sharpe_ratio = get_portfolio_stats(port_val)
    error = sharpe_ratio * -1
    return error

def find_optimal_allocations(prices):
    """Find optimal allocations for a stock portfolio, optimizing for Sharpe ratio, 
    given the daily prices for each stock in the portfolio"""
    guess = [1.0/prices.shape[1]] * prices.shape[1]
    bnds = [[0,1]] * prices.shape[1]
    cons = ({ 'type': 'eq', 'fun': lambda guess: 1.0 - np.sum(guess) })
    result = spo.minimize(error_optimal_allocations, guess, args = (prices,), method='SLSQP', bounds = bnds, 
                          constraints = cons, options={'disp':True})
    allocs = result.x
    return allocs
```

**Run optimization and disply results**

```python
# Get optimal allocations
opt = find_optimal_allocations(prices)
opt_n = opt / np.sum(opt)  # normalize allocations, if they don't sum to 1.0
allocs = dict(zip(symbols, opt_n))
print "\n", allocs

# Get daily portfolio value 
port_val = get_portfolio_value(prices, allocs.values())

# Get portfolio statistics
port_cum_ret, port_avg_daily_ret, port_std_daily_ret, port_sharpe_ratio = get_portfolio_stats(port_val)

# Plot daily portfolio value versus SPY
df_temp = pd.concat([port_val, prices_SPY / prices_SPY.ix[0]], keys=['Portfolio', 'SPY'], axis=1)
plot_data(df_temp, title="Daily Portfolio Value and SPY")

# Print statistics
print "Start Date:", date_start
print "End Date:", date_end
print "Symbols:", symbols
print "Optimal allocations:", allocs.values()
print "Sharpe Ratio:", port_sharpe_ratio
print "Volatility (stdev of daily returns):", port_std_daily_ret
print "Average Daily Return:", port_avg_daily_ret
print "Cumulative Return:", port_cum_ret
```

### Compute statistics

```python
# Normalize all stock prices
prices_n = prices / prices.ix[0]
# Compute cumulative ruturn (from beginning to end)
cum_ret = prices_n.ix[-1] - prices_n.ix[0]
# Compute daily return, average and standard deviation (volatility)
daily_returns = compute_daily_returns(prices)
avg_daily_ret = daily_returns.mean()
std_daily_ret = daily_returns.std()
# Compute sharp ratio
sharpe_ratio = compute_sharpe_ratio(prices)
# Group stocks statistics
stat_stocks = pd.concat([cum_ret, avg_daily_ret, std_daily_ret, sharpe_ratio], 
                        keys=['cum_ret', 'avg_daily_ret', 'std_daily_ret', 'sharpe_ratio'],axis=1)
# Create a dataframe for the optimized portfolio 
d = {'cum_ret': port_cum_ret, 'avg_daily_ret': port_avg_daily_ret, 
             'std_daily_ret': port_std_daily_ret, 'sharpe_ratio': port_sharpe_ratio}
stat_port = pd.DataFrame(d, index=['portfolio'])
# Group all togheter
stat = stat_stocks.append(stat_port)

# Plot
colors_array = []
for stock in stat.index:
    if stock == "portfolio":
        colors_array.append('green')
    else:
        colors_array.append('blue')

fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(stat['std_daily_ret'], stat['avg_daily_ret'], c=colors_array)

tick_all = stat.index.tolist()
for i, txt in enumerate(tick_all):
    ax.annotate(txt, (stat['std_daily_ret'][i], stat['avg_daily_ret'][i]))
    
plt.xlabel('Risk')
plt.ylabel('Return')
plt.grid()
plt.show()

stat.reindex(columns=['cum_ret', 'avg_daily_ret', 'std_daily_ret', 'sharpe_ratio'])
```
